```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("F:\\Dinesh personal\\ExcelR\\assignments\\Assignment 7\\crime_data.csv", index_col=0)
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import normalize
```

```python
df1=normalize(df)
```

```python
df1
```

```python
norm=pd.DataFrame(df1,columns=df.columns)
```

```python
norm.head()
```

```python
wcss=[]
for i in range (1,15):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,15),wcss)
```

```python
clusters=KMeans(3).fit(norm)
```

```python
clusters
```

```python
norm['cluster_3']=clusters.labels_
```

```python
norm
```
